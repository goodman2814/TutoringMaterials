## This is a guide to show how to use Jupyter notebooks with VS Code

## Start by opening VS Code
    * If you don't have it downloaded, find it at the following link
        * https://code.visualstudio.com/download

* Open the file explorer and navigate to the folder in which you will be writing

* Right-click on the notebook that you wish to open
* A new menu with open
    * select 'Open in Notebook Editor'

* Your notebook will open in a new 'tab' in VS Code
* You can now begin to edit the notebook and execute the cells
