Guide to using Jupyter Lab

* Open the computer's command prompt
    * On Windows, search for 'cmd'
    * On Mac search for 'terminal'

* In your terminal, check for your version of pip by entering the following command
    * pip --version
* As long as you don't receive an error, you should be fine to proceed

* Next check to see if you have JupyterLab installed by entering the following command
    * jupyter lab --version

* If you have a version installed, you can navigate to your folder where you will be writing by using the command to change directory: cd
       * Type cd followed by a space and then the path to your folder

* Once you have navigated to your chosen directory, type: jupyter lab 
* This should trigger your browser to create a new tab, and you can begin to write code
